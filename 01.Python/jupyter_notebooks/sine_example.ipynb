```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3], [8, 12, -1])
```

```python
import numpy as np
plt.plot(np.array([1, 2, 3]), np.array([8, 12, -1]), c='red')
```

```python
X = np.linspace(0, 20, 201)
Y = np.sin(X)
plt.plot(X,Y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X.shape, Y.shape
X = X.reshape(-1, 1)
X.shape
model.fit(X, Y)
model.coef_
```

```python
Y_pred = model.predict(X)
```

```python
plt.plot(X, Y_pred)
```

```python
linear_score = model.score(X, Y)
print(f'the score is {linear_score}')
```

Now we want to try a _polynomial regression_

```python
from sklearn.preprocessing import PolynomialFeatures
my_new_features = PolynomialFeatures(4)
my_new_features.fit(X)
expanded_X = my_new_features.transform(X)
expanded_X.shape
```

```python
model = LinearRegression()

model.fit(expanded_X, Y)
model.coef_
```

```python
Y_pred_poly = model.predict(expanded_X)
plt.plot(X, Y_pred_poly, X, Y)
```

We used the separate steps of the process manually, but it's much nicer to use scikit make_pipeline to do that

```python
from sklearn.pipeline import make_pipeline
model = make_pipeline(PolynomialFeatures(5), PolynomialFeatures(2), LinearRegression())
model.fit(X, Y)
predicted_Y = model.predict(X)
plt.plot(X, predicted_Y, X, Y)
```
