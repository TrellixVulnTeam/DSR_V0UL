```python
from  matplotlib  import pyplot
from keras import backend as K
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# read and plot image using mpimg (image path: data_aug/meerkat.jpg)

# add your implementation
img_path = 'data_aug/meerkat.jpg'
#mpimg.(mpimg.imread(img_path))
img = mpimg.imread(img_path)
plt.imshow(img)
```

```python
import numpy as np
X_train = np.array([img])
y_train = np.array([1.0])
```

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(zoom_range=3)
```

```python
for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=1):
    plt.subplot()
    plt.imshow(np.array(X_batch[0], dtype="uint8"))
    plt.show()
    break
```

```python
# use ImageDataGenerator from keras.preprocessing.image to generate different variations of the above image
import keras
# add your implementation

datagen = keras.preprocessing.image.ImageDataGenerator(
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True)
```

```python
x = datagen.get_random_transform(img.shape, seed=None)
print(x)
plt.imshow(datagen.apply_transform(img, x))
```

```python
from ipywidgets import interact

@interact(x=(0, 90))
def rotate(x):
    plt.imshow(datagen.apply_transform(img, {'theta': x}))
```

```python
@interact(x=(0,900))
def translate(x):
    plt.imshow(datagen.apply_transform(img, {'tx': x}))
```

```python
@interact(x=(0, 800))
def translate(x):
    plt.imshow(datagen.apply_transform(img, {'ty': -x}))
```
