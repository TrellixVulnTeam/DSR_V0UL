```python
from csv import DictReader

number_of_reviews = []
review_scores_value = []

for apartment in DictReader(open('listings.csv')):
    number_of_reviews.append(apartment['number_of_reviews'])
    review_scores_value.append(apartment['review_scores_value'])

review_scores_value
```

```python
import numpy as np
number_of_reviews = np.array(number_of_reviews)
review_scores_value = np.array(review_scores_value)
number_of_reviews
```
